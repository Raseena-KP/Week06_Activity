# 2. Find out how many countries with respect to the region are available in the dataset

```python
df3=df["Region"].value_counts()
df3=pd.DataFrame(df3)
df3.reset_index (inplace=True) 
df3.columns=["Region","No. of Countries"]
df3
```

# 3. On average which region has the highest
    a. GDP per capita (current US$)
    b. International trade: Imports (million US$)

```python
df3=df
x="GDP_per_capita_(current_US$)"
y="International_trade:_Imports_(million_US$)"
pt_1=df3.pivot_table(index="Region",values=[x,y],aggfunc=statistics.mean)
pt_1.sort_values(by=[x,y],ascending=False)
```

Remark: On average NorthernAmerica region has the highest GDP per capita (current US$)& International trade: Imports (million US$)

# 4. Find out which region has more consistent with respect to GDP per capita 
(current US$)

```python
df4=df
z="GDP_per_capita_(current_US$)"
pt_2=df4.pivot_table(index="Region",values=z,aggfunc=statistics.variance)
pt_2.sort_values(by=z,ascending=True)
```

Remark: WesternAfrica has more consistent with respect to GDP per capita(Less Variance)
